## Multiclassification

Contrary to the pets dataset studied in last lesson, here each picture can have multiple labels. If we take a look at the csv file containing the labels (in 'train_v2.csv' here) we see that each 'image_name' is associated to several tags separated by spaces.

```python
df = pd.read_csv(path/'train_v2.csv')
df.head()
```

\# csv 읽어와서
head로 초반부만 보여줌

To put this in a `DataBunch` while using the [data block API](https://docs.fast.ai/data_block.html), we then need to using `ImageList` (and not `ImageDataBunch`). This will make sure the model created has the proper loss function to deal with the multiple classes.

```python
tfms = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)
```

Q? 지난번에도 사용했는데 정확히 모르겠어요

> ## doc(get_transforms) 한 번 써보겠어요??
> ## get_transforms의 인자들, 인풋들, 그리고 아웃풋을 한 번 생각해보세요

We use parentheses around the data block pipeline below, so that we can use a multiline statement without needing to add '\\'.

```python
np.random.seed(42)
src = (ImageList.from_csv(path, 'train_v2.csv', folder='train-jpg', suffix='.jpg')
       .split_by_rand_pct(0.2)
       .label_from_df(label_delim=' '))
```

\# 랜덤시드 설정 및 이미지 리스트 받아오기

```python
data = (src.transform(tfms, size=128)
        .databunch().normalize(imagenet_stats))
```

\# 이미지 데이터 번치를 생성

`show_batch` still works, and show us the different labels separated by `;`.

```python
data.show_batch(rows=3, figsize=(12,9))
```

\# 3x3, 가로 12 세로9로 보여주기 (아직 멀티 레이블이 남아있음 ;)

To create a `Learner` we use the same function as in lesson 1. Our base architecture is resnet50 again, but the metrics are a little bit differeent: we use `accuracy_thresh` instead of `accuracy`. In lesson 1, we determined the predicition for a given class by picking the final activation that was the biggest, but here, each activation can be 0. or 1. `accuracy_thresh` selects the ones that are above a certain threshold (0.5 by default) and compares them to the ground truth.

As for Fbeta, it's the metric that was used by Kaggle on this competition. See [here](https://en.wikipedia.org/wiki/F1_score) for more details.

```python
arch = models.resnet50
```

\# 러닝 모델을 arch로 받아옴

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
f_score = partial(fbeta, thresh=0.2)
learn = cnn_learner(data, arch, metrics=[acc_02, f_score])
```

\# 러너를 생성

정확도와 F베타 값을 결과에 포함

```python
help(partial)
```

We use the LR Finder to pick a good learning rate.

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

\# 최적의 학습율을 찾아 보여줌

Then we can fit the head of our network.

```python
lr = 0.01
```

\# 학습율 0.01로 설정

```python
learn.fit_one_cycle(5, slice(lr))
```

\# 5회 학습

```python
learn.save('stage-1-rn50')
```

\# 학습결과 저장

...And fine-tune the whole model:

```python
learn.unfreeze()
```

\# 러너 재가동

```python
learn.lr_find()
learn.recorder.plot()
```

\# 최적값 탐색

```python
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

\# 5회 학습

```python
learn.save('stage-2-rn50')
```

\# 학습결과 저장

```python
data = (src.transform(tfms, size=256)
        .databunch().normalize(imagenet_stats))

learn.data = data
data.train_ds[0][0].shape
```

\# 새로운 데이터 번치를 만들어서 러너의 데이터에 덮어 씌움. 이후 학습 모델의 첫번째 데이터의 크기 출력

```python
learn.freeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

\# 러너 재가동하여 최적 경로 찾기

```python
lr=1e-2/2
```

\# 학습율 재설정

```python
learn.fit_one_cycle(5, slice(lr))
```

\# 5번 학습

```python
learn.save('stage-1-256-rn50')
```

\# 학습 결과 저장

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

\# 재가동하고 5회 반복

```python
learn.recorder.plot_losses()
```

\# 손실 출력

```python
learn.save('stage-2-256-rn50')
```

```python
learn = load_learner(path, test=test)
preds, _ = learn.get_preds(ds_type=DatasetType.Test)
```

\# 러너를 로드하고, test에서 정보를 받아옴

```python
thresh = 0.2
labelled_preds = [' '.join([learn.data.classes[i] for i,p in enumerate(pred) if p > thresh]) for pred in preds]
```

\# 쓰레드를 설정

Q? preds의 기능을 모르겠어요

```python
labelled_preds[:5]
```

\# preds 5개 출력

```python
fnames = [f.name[:-4] for f in learn.data.test_ds.items]
```

\# 러너의 항목에서 뒤에 4글자만 따서 이름을 지음

```python
df = pd.DataFrame({'image_name':fnames, 'tags':labelled_preds}, columns=['image_name', 'tags'])
```
